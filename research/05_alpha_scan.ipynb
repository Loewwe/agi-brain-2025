# Stage 8 Alpha Scan — Batch 001 Results

Analysis of 6 experiments: Momentum, Vol Expansion, Reversal × BTC/ETH.

**Setup:**
- Train: 2024-01-01 to 2024-03-31
- Test: 2024-04-01 to 2024-04-30
- Features: Extended (OBV, Multi-TF, Microstructure)
- Model: LightGBM

```python
import json
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (12, 6)
```

## 1. Load All Results

```python
# Load all experiment results
results = []
for p in sorted(Path('../results').glob('exp_*.json')):
    with open(p) as f:
        data = json.load(f)
        data['exp_id'] = p.stem
        results.append(data)

df = pd.DataFrame(results)

# Extract experiment details from ID
df['exp_num'] = df['exp_id'].str.extract(r'exp_(\d+)')[0].astype(int)
df['target'] = df['exp_id'].str.extract(r'_(momentum|volexp|reversal)_')[0]
df['symbol'] = df['exp_id'].str.extract(r'_(btc|eth)_')[0].str.upper()

print(f"Loaded {len(df)} experiments")
df.head()
```

## 2. Summary Table

```python
# Select key

 metrics
summary = df[[
    'exp_id', 'target', 'symbol', 'auc', 'win_rate', 'sharpe', 
    'profit_factor', 'total_return', 'n_trades'
]].copy()

# Sort by Sharpe (descending)
summary = summary.sort_values('sharpe', ascending=False)

# Format for display
summary['auc'] = summary['auc'].apply(lambda x: f"{x:.3f}" if pd.notna(x) else '')
summary['win_rate'] = summary['win_rate'].apply(lambda x: f"{x:.1%}" if pd.notna(x) else '')
summary['sharpe'] = summary['sharpe'].apply(lambda x: f"{x:+.2f}" if pd.notna(x) else '')
summary['profit_factor'] = summary['profit_factor'].apply(lambda x: f"{x:.3f}" if pd.notna(x) else '')
summary['total_return'] = summary['total_return'].apply(lambda x: f"{x:+.1%}" if pd.notna(x) else '')

print("\n=== EXPERIMENT RESULTS (sorted by Sharpe) ===")
print(summary.to_string(index=False))
summary
```

## 3. Visualization: Performance Metrics

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# AUC
ax = axes[0, 0]
df.plot(kind='bar', x='exp_id', y='auc', ax=ax, color='steelblue', legend=False)
ax.set_title('AUC (ML Signal Strength)', fontsize=14, fontweight='bold')
ax.set_ylabel('AUC')
ax.axhline(0.5, color='red', linestyle='--', alpha=0.5, label='Random')
ax.legend()
ax.set_xticklabels(df['exp_id'], rotation=45, ha='right')

# Sharpe Ratio
ax = axes[0, 1]
colors = df['sharpe'].apply(lambda x: 'green' if x > 0 else 'red')
df.plot(kind='bar', x='exp_id', y='sharpe', ax=ax, color=colors, legend=False)
ax.set_title('Sharpe Ratio (Risk-Adjusted Return)', fontsize=14, fontweight='bold')
ax.set_ylabel('Sharpe')
ax.axhline(0, color='black', linestyle='-', alpha=0.3)
ax.set_xticklabels(df['exp_id'], rotation=45, ha='right')

# Profit Factor
ax = axes[1, 0]
colors = df['profit_factor'].apply(lambda x: 'green' if x > 1 else 'red')
df.plot(kind='bar', x='exp_id', y='profit_factor', ax=ax, color=colors, legend=False)
ax.set_title('Profit Factor', fontsize=14, fontweight='bold')
ax.set_ylabel('PF')
ax.axhline(1.0, color='black', linestyle='-', alpha=0.3, label='Break-even')
ax.legend()
ax.set_xticklabels(df['exp_id'], rotation=45, ha='right')

# Total Return
ax = axes[1, 1]
colors = df['total_return'].apply(lambda x: 'green' if x > 0 else 'red')
df.plot(kind='bar', x='exp_id', y='total_return', ax=ax, color=colors, legend=False)
ax.set_title('Total Return (Test Period)', fontsize=14, fontweight='bold')
ax.set_ylabel('Return')
ax.set_yticklabels([f'{y:.0%}' for y in ax.get_yticks()])
ax.axhline(0, color='black', linestyle='-', alpha=0.3)
ax.set_xticklabels(df['exp_id'], rotation=45, ha='right')

plt.tight_layout()
plt.savefig('../results/alpha_scan_summary.png', dpi=150, bbox_inches='tight')
plt.show()

print("\nVisualization saved to: results/alpha_scan_summary.png")
```
